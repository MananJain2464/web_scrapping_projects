```python
import requests
```

```python
link = 'https://quotes.toscrape.com/'
```

```python
res = requests.get(link)
html = res.text 
```

```python
print(html)
```

```python
with open('main.html', 'w', encoding='utf-8') as fd:
    fd.write(html)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html , 'html.parser')

```

```python
print(soup.find('span' , class_ = 'text').text[1:-1])
```

```python
soup.find_all('span' , class_ = 'text')
```

```python
quotes = []

for quote in soup.find_all('span' , class_ = 'text'):
    quotes.append(quote.text[1:-1])

quotes
```

```python

quotes_info = []

for i in soup.find_all('div' , class_='quote'):
    quote = i.find('span' , class_='text').text[1:-1]
    author = i.find('small' , class_='author').text
    author_link = i.find('a').get('href')
    tags = []
    for j in i.find_all('a' , class_ = 'tag'):
        tags.append(j.text)
    
    tags = ','.join(tags)
    
    quotes_info.append([quote , author , author_link , tags])
    
    print(quote)
    print(author)
    print(author_link)
    print(tags)
    
    print('*'*40)
        
    
```

```python
quotes_info
```

```python
import pandas as pd
```

```python
my_data = pd.DataFrame(quotes_info , columns = ['quote' , 'author' , 'author_id' , 'tags'])
my_data
```

```python
my_data.to_csv('Quotes.csv')
```

```python
from tqdm import tqdm 

for k in tqdm(range(2,11)):
    link = 'https://quotes.toscrape.com/page/'+str(k)
    my_link = requests.get(link)
    soup = BeautifulSoup(my_link.text , 'html.parser')
    
    for i in soup.find_all('div' , class_='quote'):
        quote = i.find('span' , class_='text').text[1:-1]
        author = i.find('small' , class_='author').text
        author_link = i.find('a').get('href')
        tags = []
        for j in i.find_all('a' , class_ = 'tag'):
            tags.append(j.text)
        
        tags = ','.join(tags)
        
        quotes_info.append([quote , author , author_link , tags])
        
        # print(quote)
        # print(author)
        # print(author_link)
        # print(tags)
        
        # print('*'*40)
        
```
