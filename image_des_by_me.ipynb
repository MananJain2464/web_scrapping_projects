```python
import requests

from selenium import webdriver 

from bs4 import BeautifulSoup

link = 'https://stock-pictures.netlify.app'
```

```python
browser = webdriver.Chrome()
```

```python
browser.get(link)
```

```python
from tqdm import tqdm 
```

```python
import time 

for i in tqdm(range(0 , 10000000, 10000)):
    
    browser.execute_script('window.scrollTo (0 ,'+ str(i) + ')')
    time.sleep(.1)
    
```

```python
browser.page_source
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
sp = soup.find_all('div' , class_='container')

data = []
for x in tqdm(sp):
    image_link = x.find('img' , class_='source-img').get('src')
    tag = list(set(x.find('span' , class_='tag-color').text[7:].split(', ')))
    y =   x.find('div' , class_='likes-comments')
    likes = y.find_all('span')[0].text[:-5]
    comments = y.find_all('span')[1].text [:-9]
    
    data.append([image_link , tag , likes , comments])
    
```

```python
len(data)
```

```python
import pandas as pd 

my_data = pd.DataFrame(data = data , columns=['image_link' , 'tag' , 'likes' , 'comments'])
my_data.head()
```

```python
my_data.to_csv('images_csv.csv')
```

```python
link = my_data['image_link'][0]

res = requests.get(link)

fd = open('Img.png' , 'wb')

fd.write(res.content)

fd.close()
```

```python
import os
import requests
from tqdm import tqdm

# Ensure the directory exists
if not os.path.exists('Imgs'):
    os.makedirs('Imgs')

for link in tqdm(my_data['image_link']):
    res = requests.get(link)
    
    # Generate the file path
    file_path = os.path.join('Imgs', link.split('/')[-1])
    
    # Write the content to the file
    with open(file_path, 'wb') as fd:
        fd.write(res.content)
```
