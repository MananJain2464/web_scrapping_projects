```python
import requests 
from bs4 import BeautifulSoup
```

```python
title = str(input('Enter the topic :')).replace( ' ' , '+')

link = 'https://www.google.com/search?q='+ title + '+wikipedia'
print(link)
```

```python
google_res = requests.get(link)
google_soup = BeautifulSoup(google_res.text , 'html.parser')

for i in google_soup.find_all('div'):
    try:
        final_link = i.find('a').get('href')
        if 'en.wikipedia.org' in final_link:
            #print(final_link)
            break
    except:
        pass

final_link = final_link[7:]
final_link = final_link.split('&')[0]
final_link
        
```

```python
res = requests.get(final_link)
soup = BeautifulSoup(res.text , 'html.parser')
```

```python
heading = soup.find('h1').text
print(heading)
```

```python
for i in soup.find_all('p'):
    print(i.text)
    print('-'*50)
```

```python
my_text = ''
for i in soup.find_all('p'):
    my_text += i.text
    my_text+='\n'

j=2
for j in range(2 , 500):
    my_text = my_text.replace('['+str(j) + ']' , '')

print(my_text)
```

```python
fd = open(heading+'.txt' , 'w' , encoding='utf-8')
fd.write(my_text)
fd.close
```
