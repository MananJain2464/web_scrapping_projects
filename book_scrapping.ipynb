```python
import requests

import pandas as pd 

from bs4 import BeautifulSoup

from tqdm import tqdm 
```

```python
data = []

for j in tqdm(range(1 , 51)):
    link = 'https://books.toscrape.com/catalogue/page-'+str(j) + '.html'

    res = requests.get(link)

    html = res.text

    soup = BeautifulSoup(html , 'html.parser')

    for i in soup.find_all('li' , class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):
        
        book_name = i.find_all('a')[-1].get('title')
        book_link = 'https://books.toscrape.com/catalogue/' + i.find_all('a')[-1].get('href')
        image_link = 'https://books.toscrape.com/'+i.find('img').get('src')
        book_rating = i.find('p').get('class')[-1]
        price = i.find('p' , class_ = 'price_color')
        availability = i.find('p' , class_='instock availability').text.strip()

        data.append([book_name , book_link , image_link , book_rating , price , availability])


    
    
```

```python
len(data)
```

```python
my_data = pd.DataFrame(data , columns=['book_name' , 'book_link' , 'image_link' , 'book_rating' , 'price' , 'availability'])
```

```python
if 'price' in my_data.columns:
    # Convert all entries in 'price' to strings
    my_data['price'] = my_data['price'].astype(str)
    
    # Remove the first and last character from each entry
    my_data['price'] = my_data['price'].str[3:-3]
    
    # Display the modified DataFrame
    my_data.head()

```

```python
my_data.to_csv('book_scrapper.csv')
```

```python
data = []
for link in tqdm(my_data['book_link']):
    res= requests.get(link)
    soup = BeautifulSoup(res.text , 'html.parser')
    typ = soup.find('ul' , class_ = 'breadcrumb').find_all('a')[2].text 
    
    temp = soup.find('table' , class_='table table-striped').find_all('td')
    
    upc = temp[0].text
    price_x = temp[2].text[2:]
    price_i = temp[3].text[2:]
    tax = temp[4].text[2:]
    qn = temp[5].text
    reviews = temp[6].text
    
    data.append([typ,upc , price_x , price_i , tax , qn , reviews ])
```

```python
second_df = pd.DataFrame(data, columns = ['type' , 'upc' , 'price_x' , 'price_i' , 'tax' , 'quantity' , 'reviews'])
```
